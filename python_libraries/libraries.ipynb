```python
from tkinter import Tk, filedialog, Label
from PIL import Image, ImageTk

# Create the main window
root = Tk()
root.title("Image Viewer")

# Function to open and display image
def open_image():
    file_path = filedialog.askopenfilename(
        filetypes=[("Image files", "*.jpg *.jpeg *.png *.bmp *.gif")]
    )
    if file_path:
        img = Image.open(file_path)
        img = img.resize((300, 300))  # Resize for display
        img_tk = ImageTk.PhotoImage(img)
        label.config(image=img_tk)
        label.image = img_tk  # Keep a reference

# Add a label to display the image
label = Label(root)
label.pack()

# Open file dialog on start
open_image()

root.mainloop()
```

## 3D GRAPH


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("3D Surface Plot")
ax.set_xlabel("X axis")
ax.set_ylabel("Y axis")
ax.set_zlabel("Z axis")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Create some random 3D data
np.random.seed(0) # for reproducibility
n_points = 100
x_data = np.random.rand(n_points)
y_data = np.random.rand(n_points)
z_data = np.random.rand(n_points)
colors = np.random.rand(n_points) # for coloring points
sizes = 100 * np.random.rand(n_points) # for sizing points

# Create the 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
scatter = ax.scatter(x_data, y_data, z_data, c=colors, s=sizes, cmap='viridis', alpha=0.7)

# Add labels and title
ax.set_xlabel('A Axis')
ax.set_ylabel('B Axis')
ax.set_zlabel('C Axis')
ax.set_title('Scatter Plot Graph in 3D')

# Add a color bar
fig.colorbar(scatter, label='Color Intensity')

# Show the plot
plt.show()
```
